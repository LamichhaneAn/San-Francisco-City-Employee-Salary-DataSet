18. FInd the Average BasePay of All Employee  per year

```python
import pandas as pd

# Load your data
data = pd.read_csv("Salaries.csv")

# Replace "Not Provided" with NaN and convert 'BasePay' to float
data['BasePay'] = data['BasePay'].replace("Not Provided", np.nan).astype(float)

# Group by 'Year' and calculate the mean 'BasePay' for each year
average_basepay_per_year = data.groupby("Year")["BasePay"].mean()

print(average_basepay_per_year)
```

19. Find the Average BasePay of all Employee per Job title

```python
import pandas as pd
import numpy as np

# Load your data
data = pd.read_csv("Salaries.csv")

# Replace "Not Provided" with NaN and convert 'BasePay' to float
data['BasePay'] = data['BasePay'].replace("Not Provided", np.nan).astype(float)

# Group by 'JobTitle' and calculate the mean 'BasePay' for each job title
average_basepay_per_jobtitle = data.groupby("JobTitle")["BasePay"].mean()

print(average_basepay_per_jobtitle)
```

20. Find the Average BasePay of Employee Having Job Title ACCOUNTANT

```python
import pandas as pd
import numpy as np

# Load your data
data = pd.read_csv("Salaries.csv")

# Replace "Not Provided" with NaN and convert 'BasePay' to float
data['BasePay'] = data['BasePay'].replace("Not Provided", np.nan).astype(float)

# Find the average BasePay for employees with the job title "ACCOUNTANT"
average_basepay_accountant = data[data["JobTitle"] == "ACCOUNTANT"]["BasePay"].mean()

print("Average BasePay for ACCOUNTANT:", average_basepay_accountant)
```
